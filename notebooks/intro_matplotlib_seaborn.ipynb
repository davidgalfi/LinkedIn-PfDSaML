```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {
    'Name': ['Steve', 'John', 'Richard', 'Sarah', 'Randy', 'Michael', 'Julie'],
    'Age': [20, 22, 20, 21, 24, 23, 22],
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'Rank': [2, 1, 4, 5, 3, 7, 6]
}
```

```python
people_df = DataFrame(data)
people_df
```

Matplotlib

```python
plt.bar(people_df['Name'], people_df['Age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

Seaborn

```python
plot = sns.barplot(data=people_df, x='Name', y='Age')
plot.set_title('Comparing Ages')
plt.show()
```

```python
plot = sns.pointplot(data=people_df, x='Name', y='Age')
plt.show()
```
